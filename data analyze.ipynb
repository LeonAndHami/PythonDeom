```python
from matplotlib import pyplot as plt
import random
font = {'family' : 'SimHei',
        'weight' : 'bold',
        'size'   : '16'}
plt.rc('font', **font)               # 步骤一（设置字体的更多属性）
plt.rc('axes', unicode_minus=False)  # 步骤二（解决坐标轴负数的负号显示问题）
```

```python
# 画折线图
plt.figure(figsize=(20,8),dpi=80) # 设置图片大小
# figure 图形图标的意思，在这里指的就是我们画的图
# 通过实例化一个figure并且传递参数，能够在后台自动使用figure实例
# 在图像模糊的时候可以传入dip参数，让图片更加清晰
x = range(2,26,2)
y = [15,13,14.5,17,20,25,26,26,24,22,18,15]
plt.plot(x,y) # 画图

# 设置x轴的刻度
plt.xticks(range(2,25))

# 设置y轴刻度
# plt.yticks(range(min(y),max(y)+1))

# plt.savefig(".sig_size.pgn") # 可以保存为svg矢量图格式，放大不会有锯齿
plt.show()
```

```python
# 显示10点到12点每分钟的温度
x = range(120)
y = [random.randint(20,35) for i in range(120)]
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y)

# matplotlib 默认不支持中文，可以通过matplotlib.rc方法来设置
font = {'family' : 'SimHei',
        'weight' : 'bold',
        'size'   : '16'}
plt.rc('font', **font)               # 步骤一（设置字体的更多属性）
plt.rc('axes', unicode_minus=False)  # 步骤二（解决坐标轴负数的负号显示问题）


# 调整x刻度
x_tick_labels = ["10点{}分".format(i) for i in range(60)]
x_tick_labels += ["11点{}分".format(i) for i in range(60)]

# 数据与label个数要一致
plt.xticks(list(x)[::3],x_tick_labels[::3],rotation=60)  # 第二个参数是label，第三个参数是旋转label的度数

# 添加描述信息
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("10点到12点每分钟的气温变化")

plt.show()
```

```python
# 11岁到30岁每年的交友数量
x = range(11, 31)
y = [random.randint(1,6) for i in range(20)]
plt.figure(figsize=(20,8),dpi=80)

# 可以用color参数设置颜色，linestyle设置线条风格，linewidth设置线条粗细，alpha设置透明度
# linestyle = '--'  虚线
plt.plot(x,y,label='我') 
plt.plot(x,[random.randint(1,6) for i in range(20)],label='小明')

x_ticks_label = ["{}岁".format(i) for i in range(11,31)]
plt.xticks(x, x_ticks_label)
plt.yticks(range(10))
plt.xlabel("年龄")
plt.ylabel("交友个数")
plt.title("11岁到30岁每年交友个数")

# 绘制网络
plt.grid()
# plt.grid(alpha=0.3) # 设置网格线透明度

plt.legend() # 显示图例 plot方法中要加 label，loc 参数可控制图例的位置


plt.show()
```

```python
# 绘制散点图
y_3 = [11,17,16,11,12,11,12,6,6,7,8,9,12,15,14,17,18,21,16,17,20,14,15,15,15,19,21,22,22,22,23,]
y_10 = [26,26,28,19,21,17,16,19,18,20,20,19,22,23,17,20,21,20,22,15,11,15,5,13,17,10,11,13,12,13,6]

x_3 = range(1, 32)
x_10 = range(51, 82) # 为了数据不叠在一起

# 设置大小
plt.figure(figsize=(20,8),dpi=80)





# 设置刻度
_x = list(x_3) + list(x_10)
_xtick_labels = ['3月{}日'.format(i) for i in x_3]
_xtick_labels += ['10月{}日'.format(i) for i in x_3]
plt.xticks(_x[::3],_xtick_labels[::3],rotation=45)


# 添加描述信息
plt.xlabel('时间')
plt.ylabel('温度')
plt.title('北京3月份和10月份每天最高气温')



plt.scatter(x_3, y_3,label='3月份')
plt.scatter(x_10, y_10,label='10月份')

plt.legend(loc=2) #loc控制显示在左上



```
